```python
plt.subplot(1,2,1)
plt.plot(x,y,"r--")
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

Subplot (1,2,1)의 의미는, 기존 plot을 1을 기준으로 가로에 두개의 그래프를 넣는다는 의미이고, 그 다음 1은 두개의 그래프 위치 중 첫번째에 plt.plot(x, y, 'r--')  라는 그래프를 넣는다는 의미이다

```python
plt.subplot(2,3,1)
plt.plot(x,y,"r--")
plt.subplot(2,3,3)
plt.plot(y,x,'g*-')
```

```python
plt.subplot(2,3,2)
plt.plot(x,y,"r--")
plt.subplot(1,3,1)
plt.plot(y,x,'g*-')
```

plt.subplot(1,3,1) 은 기존 plot의 범위에 3개의 그래프를 넣는다는 의미이고, 그 그래프 위치 3가지 중 첫번째에 plt.plot(x,y,'r--')그래프를 넣는다는 뜻이다.

```python
plt.subplot(1,4,1)
plt.plot(x,y,'y--')
plt.subplot(1,4,3)
plt.plot(y,x,'b*-')
```
